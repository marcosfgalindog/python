**COEFICIENTES DE CORRELACION**

```python
corr_OverallQual = np.corrcoef(training_data[:,0],training_data[:,1])[1,0]
corr_1stFlrSF = np.corrcoef(training_data[:,0],training_data[:,2])[1,0]
corr_TotRmsAbvGrd = np.corrcoef(training_data[:,0],training_data[:,3])[1,0]
corr_YearBuilt = np.corrcoef(training_data[:,0],training_data[:,4])[1,0]

#Se crea nuevo array de precios que no contengan las mismas filas que poseen valores nan en el array LotFrontage
new_sale_price = training_data[:,0][~np.isnan(training_data[:,5])]
corr_LotFrontage = np.corrcoef(new_sale_price,new_LotFrontage)[1,0]
```

```python
scatter_TotRmsAbvGrd = plt.scatter(training_data[:,3],training_data[:,0])
plt.title('Sale Price and Total Rooms Above Grade, coefficient: '+ str(corr_TotRmsAbvGrd))
plt.xlabel('Total Rooms Above Grade')
plt.ylabel('Sale Price')
plt.show()
```

```python
scatter_YearBuilt = plt.scatter(training_data[:,4],training_data[:,0])
plt.title('Sale Price and Year of Construction, coefficient: '+ str(corr_YearBuilt))
plt.xlabel('Year of Construction')
plt.ylabel('Sale Price')
plt.show()
```
