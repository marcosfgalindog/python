**FUNCION PARA ENTRENAMIENTO DE MODELO LINEAL y=mx+b**

```python
#Funcion para entrenamiento del modelo lineal
OverAll_dataset = np.reshape(training_data[:,1],(1168,1))
FrstFlrSF_dataset = np.reshape(training_data[:,2],(1168,1))
SalePrice_dataset = np.reshape(training_data[:,0],(1168,1))


def training(variable_x, variable_y, epochs, imprimir_cada_error, learning_rate):
    
    m_b = np.array([[0,0]]).transpose()
    errores_por_iteracion = np.array([]) 
    variable_y = np.reshape(variable_y, (len(variable_y),1))
    modelo_resultante = {}
    
    columna_a_crear = np.ones_like(variable_x)
    variable_matrix = np.hstack((variable_x, columna_a_crear)) #esto crea la matriz con 1's anexados a la derecha de la variable
    
    for i in range(epochs):
                
        
        y_hat = np.matmul(variable_matrix,m_b)
        pre_error =  y_hat - variable_y 
        error = np.mean(pre_error**2 /2)
        error = np.array([error])
        errores_por_iteracion = np.concatenate([errores_por_iteracion,error])
        pre_gradient = pre_error*variable_matrix
        gradient = np.mean(pre_gradient,axis=0).reshape(2,1)
        m_b =  m_b - learning_rate*gradient
        modelo_resultante[i+1] = m_b.reshape(1,2)
        
        if i % imprimir_cada_error == 0:
            print("Error No. "+str(i)+": "+ str(error)) #aca se imprime cada error que uno desee
    
    return modelo_resultante, errores_por_iteracion.transpose()
```

Funcion que genere las predicciones para cada variable seleccionada (esta puede ser usada por cada variable):

```python
def regression_results(modelo_artesanal,modelo_scikit,x):#Esta funcion puede ser usada por ambos modelos
    x_1 = np.ones_like(x)
    x_artesanal = np.vstack((x,x_1)).T
    x = x.reshape(-1,1)
    r1 = np.matmul(x_artesanal,modelo_artesanal[0][len(modelo_artesanal[0])].T)
    r2 = modelo_scikit.predict(x)
    return r1, r2, (r1+r2)/2
```
